```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

#1 - how many people are normal
#2 - how many people are pre-dibatic
#3 - how many are dibatic

# 80-100 normal
# 100-125 pre
# 125 + diabtic
```

```python
plt.hist(blood_sugar, bins=15) # bins by default number is 10
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150] , rwidth=0.70) # bins by default number is 10
```

```python
# how many are normal - 6
# how many are prediabric- 2
# how many are diabtic = 4
```

# horizontal histogram

```python
plt.hist(blood_sugar, bins=15 , orientation='horizontal') # bins by default number is 10
```
