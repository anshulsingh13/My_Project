```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,53,48,47,49]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)

#Axis label
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('weather')
```

```python
plt.plot(days, max_t, label ='max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

#Axis label
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('weather')
plt.legend()
```

```python
plt.plot(days, max_t, label ='max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

#Axis label
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('weather')
plt.legend(loc = 'best', shadow = 'True')
```

```python
plt.plot(days, max_t, label ='max')
plt.plot(days, min_t, label = 'min')
plt.plot(days, avg_t, label = 'avg')

#Axis label
plt.xlabel('week days')
plt.ylabel('temp')
plt.title('weather')
plt.legend(loc = 'best', shadow = 'True', fontsize = 'small')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]

#1 how many people are normal?
#1 how many people are pre-diabetic?
#1 how many people are diabetic?

plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=20)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150])
```

```python
#1 how many people are normal? 6
#1 how many people are pre-diabetic? 2
#1 how many people are diabetic? 4
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth = 0.70)
```

```python
plt.hist(blood_sugar, bins=[80,100], rwidth = 0.70, label = 'normal')
plt.hist(blood_sugar, bins=[100,125], rwidth = 0.70, label = 'pre')
plt.hist(blood_sugar, bins=[125,150], rwidth = 0.70, label = 'diabetic')
plt.legend(loc = 'best', shadow = 'True', fontsize = 'small')
plt.xlabel('sugar level')
plt.ylabel('frequency')
plt.title('Frequnecy of people wrt blood sugar level')
```

```python
#horizaontL HISTOGRAM

plt.hist(blood_sugar, bins=15, orientation ='horizontal')
```
