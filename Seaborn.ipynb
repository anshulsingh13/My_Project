```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# relplot can actulayy find the rrealtion between max of 5 col at a time
sns.relplot(x='total_bill' , y = 'tip' , data = tips )
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data = tips , hue='day' )
```

```python
sns.relplot(x='total_bill' , y = 'tip' , data = tips , hue='time' , style='day' , size='sex')
```

```python
sns.scatterplot(x='total_bill' , y = 'tip' , data = tips , hue='time' , style='day' , size='sex')
```

```python
iris = sns.load_dataset('iris')
```
