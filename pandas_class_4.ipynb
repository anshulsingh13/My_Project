# how to iter in pandas

```python
# 1 . itertuples
# 2 . iterrows
# 3 . iteritems
```

```python
#import lib
import pandas as pd
import numpy as np
```

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
org_df = pd.DataFrame(space_org)
```

```python
org_df
```

### 1. iter tuples

```python
for row in org_df.itertuples():
    print(row)
```

### 2. iterrows

```python
for row in org_df.iterrows():
    print(type(row[1]))
```

```python
for row in org_df.iterrows():
    print(row[1]['Country_name'])
```

```python
for abc , row in org_df.iterrows():
    print(abc)
    print(row)
```

```python
for value in  org_df.iteritems():
    print(value)
```

```python
for key , value in  org_df.iteritems():
    print(f'key value {key}')
    print(value)
```

# concatenation

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3'],
                        'E': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[0, 1, 2, 3])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1 , df2 , df3] , axis = 1 ).fillna(value = 'helo')
```

```python
a = pd.concat([df1 , df2 , df3] , axis = 0 ).fillna(value = 'helo')
```

```python
a
```
