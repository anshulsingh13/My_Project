```python
# % of missing value in cabin

c = round(df.Cabin.isnull().sum()/df.shape[0]*100,3)

# % of missing value in embarked

w = round(df.Embarked.isnull().sum()/df.shape[0]*100,3)

# % of missing value in age

a = round(df.Age.isnull().sum()/df.shape[0]*100,3)

print(c,w,a)
```

```python
#cabin  - remoe column
#embarked - mode
#age - mean/median

# df['Age'].fillna(df['Age'].median, inplace = True)
df['Embarked'].fillna(df.Embarked.value_counts().idxmax(), inplace = True)
# df.drop('Cabin', axis=1, inplace = True)
```

```python
# find common boarding point
df.Embarked.value_counts().idxmax()
```

```python
#create column with name travel alone and put value 0 if family is more than 0 else 1

df['Alone'] = df['family'].apply(lambda x : 0 if(x > 0) else 1)
df
```
